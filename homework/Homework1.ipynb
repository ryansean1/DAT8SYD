```python
# read 'u.user' and asign to 'users' so that we can re-use the data#
import pandas as pd
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_table('C:/Users/seanr/Desktop/GeneralAssembly/DAT8SYD/homework/u.user', sep='|', header=None, names=user_cols, index_col='user_id', dtype={'zip_code':str})
print (users)
```

```python
#Summarise the data (just for practice)#
users.describe()
```

```python
# for each occupation in 'users', count the number of occurrences
users.occupation.value_counts()
```

```python
#Overall users age#
users.age.mean()
```

```python
# for each occupation, calculate the mean a#
users.groupby('occupation').age.mean()

```

```python
#for each occupation, calculate the minimum and maximum ages#
users.groupby('occupation').age.agg(['min','max'])
```

```python
#for each combination of occupation and gender, calculate the mean #
users.groupby(['occupation','gender']).age.mean()
```
